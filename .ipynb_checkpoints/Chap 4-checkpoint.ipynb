<h4>4.1.1 퍼셉트론 </h4>

이진 분류(binary classification)은 임의의 샘플 데이터를 True나 False로 구분하는 문제이다.<br>
퍼셉트론은 선형 회귀와 유사한 구조를 가지지만, 마지막 단계에서 이진 분류를 위한 계단 함수(step function)를 사용한다.<br><br>

(선형 함수) w₁x₁+ w₂x₂+ b = z<br>
계단 함수는 z가 0보다 크거나 같으면 1로, 0보다 작으면 -1로 분류한다.<br>
이때 1을 양성 클래스(positive class), -1을 음성 클래스(negative class)라고 부른다.<br><br>

특성이 n개인 선형 함수를 표현하면, <br>
z = w₁x₁+ w₂x₂+ ... + w_n*x_n + b <br>
이며, 주로 시그마 기호를 사용하여 표현한다.<br>
z = b + Σw_i*x_i<br>

바로 이 함수를 활성화 함수(activation function)라고 부른다.<br>

활성화 함수는 비선형 함수를 사용한다.<br>
예를 들어,<br>
p = 1 / (1+e^-z)<br>
비선형 함수를 사용하는 이유는 만약 선형함수를 사용한다면, 선형함수를 통과해 선형함수를 사용하는 것이므로,<br>
정리하면 다시 선형함수가 나오게 된다.<br>
이는 임계 함수 앞에 뉴런을 여러 개 쌓아도 결국 선형 함수일 것이므로 의미가 없게 되기 때문에 비선형 함수를 사용한다.<br>
로지스틱 회귀의 활성화 함수로는 "시그모이드 함수"를 사용한다.


<h1>4.2 시그모이드 함수 </h1>

로지스틱 회귀의 시그모이드 함수는 선형함수의 z 값을 0~1 사이의 확률값으로 변환시켜준다.<br><br>

[오즈비(OR; Odds Ratio)]<br>
성공확률과 실패 확률의 비율을 나타내는 통계<br>
OR = p / (1-p)  ;p는 성공확률<br><br>

[로짓함수(logit function)]<br>
오즈 비에 로그 함수를 취하여 만든 함수<br>
logit(p) = log(p / (1-p))<br>
p가 0.5일 때 0이 되고, p가 0과 1일 때 각각 무한대로 음수와 양수가 된다.<br>

<h1>4.3 로지스틱 손실 함수를 경사하강법에 적용 </h1>





로지스틱 회귀와 같은 분류의 목표는 올바르게 분류된 샘플 데이터의 비율 자체를 높이는 것임.<br>
하지만, 올바르게 분류된 샘플의 비율은 미분 가능한 함수가 아니기 때문에 경사하강법을 사용할 수 없음.<br>
대신 비슷한 목표를 달성할 수 있는 함수인 "로지스틱 손실 함수"를 사용함

<h3>4.3.1 로지스틱 손실 함수</h3>

로지스틱 손실 함수는 다중 분류를 위한 손실 함수인 크로스 엔트로피(cross entropy) 손실 함수를 이진 분류 버전으로 만든 것임.<br>
a는 활성화 함수가 출력한 값이고, y는 타깃임.<br>
L = -(ylog(a) + (1-y)log(1-a))<br>
위 식은 y가 그렇다(1), 아니다(0)의 경우로 나뉜다.<br>
<li>y가 1인 경우(양성 클래스) : -log(a)</li>
<li>y가 0인 경우(음성 클래스) : -log(1-a)</li>

앞의 두 식을 최소로 만들다 보면 a의 값이 우리가 원하는 목표치가 됨. <br>
예를 들어, 양성 클래스인 경우 로지스틱 손실 함수의 값을 최소로 만들려면 a는 자연스레 1에 가까워짐.<br>
반대로 음성 클래스인 경우 로지스틱 손실 함수의 값을 최소로 만들면 a는 자연스레 0에 가까워짐.<br>

즉, 로지스틱 손실 함수를 최소화하면 a의 값이 우리가 가장 이상적으로 생각하는 값이 되므로, <br>
로지스틱 손실 함수의 최솟값을 만드는 가중치와 절편을 찾기 위해 미분을 진행하면 됨.

<h3>4.3.2 로지스틱 손실 함수 미분하기</h3>

δL/δw_i = -(y-a)x_i<br>
δL/δb = -(y-a)1<br>

미분 결과를 보면 ŷ이 a로 바뀌었을 뿐 제곱 오차를 미분한 결과와 동일함.<br>

[가중치 업데이트 방법 정리]<br>
로지스틱 회귀의 가중치 업데이트는<br>
로지스틱 손실 함수를 가중치에 대해 미분한 식을 가중치에서 빼면 됨.<br>
w_i = w_i - δL/δw_i = w_i + (y-a)x_i<br>

[절편 업데이트 방법 정리]<br>
로지스틱 손실 함수를 절편에 대해 미분한 식을 절편에서 빼면 됨.<br>
b = b - δL/δb = b + (y-a)1

<h3>4.5.2 로지스틱 회귀 구현</h3>

로지스틱 회귀는,<br>
정방향으로 데이터가 흘러가는 과정과<br>
역방향으로 가중치를 업데이트 하기 위해 데이터가 흘러가는 과정을 구현해야 함.

```python
class LogisticNeuron :
    def __init__(self):
        self.w = None
        self.b = None
    def forpass(self, x):
        z = np.sum(x * self.w) + self.b   # 직선 방정식 계산
        return z
    def backprop(self, x, err):
        w_grad = x * err
        b_grad = 1 * err
        return w_grad, b_grad
```

forpass의 np.sum(x * self.w)를 보면,<br>
x와 w가 모두 다차원 numpy 배열이므로,<br>
a = np.array([1, 2, 3]) <br>
b = np.array([3, 4, 5]) <br>
인 경우, a * b = [3, 8, 15] 이며, <br>
np.sum(a \* b) = 26이 나오게 된다.

<h3>4.5.3 훈련하는 메서드 구현</h3>

1. fit() 메서드 구현

```python
class LogisticNeuron(LogisticNeuron) :
    def fit(self, x, y, epochs=100):
        self.w = np.ones(x.shape[1])
        self.b = 0
        for i in range(epochs) :
            for x_i, y_i in zip(x, y):
                z = self.forpass(x_i)
                a = self.activation(z)
                err = -(y_i - a)
                w_grad, b_grad = self.backprop(x_i, err)
                self.w -= w_grad
                self.b -= b_grad
```

2. activation() 메서드 구현하기

```python
class LogisticNeuron(LogisticNeuron) :
    def activation(self, z):
        z = np.clip(z, -100, None) # 안전한 np.exp() 계산을 위해
        a = 1 / (1 + np.exp(-z))   # 시그모이드 계산
        return a
```

<h3>4.5.4 예측하는 메서드 구현하기</h3>

샘플에 대한 예측값을 계산할 때 forpass() 메서드를 사용함.<br>
그러나 여러 개의 샘플을 한꺼번에 예측하려면 forpass() 메서드를 여러 번 호출하게 되므로 번거로움.<br>
분류에서는 활성화 함수와 임계 학수도 적용해야 하므로 새로운 샘플에 대한 예측값을 계산해주는 predict() 메서드를 만듦

```python
class LogisticNeuron(LogisticNeuron) :
    def predict(self, x):
        z = [self.forpass(x_i) for x_i in x]   # 선형 함수 적용
        a = self.activation(np.array(z))        # 활성화 함수 적용
        return a > 0.5                         # 계단 함수 적용
```

<h1>4.6 로지스틱 회귀 뉴런으로 단일층 신경망을 만듦</h1>

위의 LogisticNeuron 클래스와 거의 비슷하지만, 손실 함수의 결괏값 저장 기능을 추가하는 단일층 신경망을 만듦.<br>

[손실 함수의 결괏값 저장 기능 추가하기]<br>
\_\_init__() 메서드에 손실 함수의 결괏값을 저장할 리스트 self.losses를 만듦.<br>
샘플마다 손실 함수를 계산하고 그 결괏값을 모두 더한 다음 샘플 개수로 나눈 평균값을 self.losses 변수에 저장.<br>

self.activation() 메서드로 계산한 a는 np.log()의 계산을 위해 한번 더 조정함.<br>
왜냐하면 a가 0에 가까워지면 np.log() 함수의 값은 무한대가, a가 1에 가까워 지면 np.log() 함수의 값은 0이 되기 때문임.<br>
손실값이 무한해지면 정확한 계산을 할 수 없으므로 a의 값이 -1x10^-10 \~ 1 - 1x10^-10 사이가 되도록 np.clip() 함수로 조정.<br>
np.clip() 함수는 주어진 범위 밖의 값을 범위 양 끝의 값으로 잘라냄

```python
class SingleLayer :
    def __init__(self):
        self.w = None
        self.b = None
        self.losses = []
    
    def forpass(self, x):
        z = np.sum(x * self.w) + self.b # 직선 방정식 계산
        return z
    
    def backprop(self, x, err):
        w_grad = x * err                # 가중치에 대한 그래디언트 계산
        b_grad = 1 * err                # 절편에 대한 그레디언트 계산
        return w_grad, b_grad
    
    def activation(self, z):
        z = np.clip(z, -100, None)     # 안전한 np.exp() 계산을 위해
        a = 1 / (1 + np.exp(-z))
        return a
    
    def fit(self, x, y, epoch=100):
        self.w = np.ones(x.shape[1])
        self.b = 0                      # 절편 초기화
        for i in range(epoch):          # epoch 만큼 반복
            loss = 0
            indexes = np.random.permutation(np.arange(len(x)))   # 인덱스를 섞음
            for i in indexes :             # 모든 샘플에 대해 반복
                z = self.forpass(x[i])     # 정방향 계산
                a = self.activation(z)     # 활성화 함수 적용
                err= -(y[i] - a)           # 오차 계산
                w_grad, b_grad = self.backprop(x[i], err) # 역방향 계산
                self.w -= w_grad
                self.b -= b_grad
                a = np.clip(a, 1e-10, 1-1e-10)
                loss += -(y[i]*np.log(a) + (1-y[i])*np.log(1-a))   # 에포크마다 평균 손실 저장
            self.losses.append(loss/len(y))
```

[여러가지 경사 하강법]<br>
<li>확률적 경사 하강법(stochastic gradient descent)</li>
샘플 데이터 1개에 대한 그래디언트를 계산<br>
계산 비용이 적지만 가중치가 최적값에 수렴하는 과정이 불안정
<li>배치 경사 하강법(batch gradient descent)</li>
전체 훈련 세트를 사용하여 한 번에 그레디언트를 계산<br>
가중치가 최적값에 수렴하는 과정은 안정적이지만 계산 비용이 많이듦
<li>미니 배치 경사 하강법</li>
배치 크기를 작게하여(훈련 세트를 여러 번 나누어) 처리하는 방식<br>
확률적 경사 하강법과 배치 경사 하강법의 중간 정도

[매 에포크마다 훈련 세트의 샘플 순서 섞어 사용하기]<br>
모든 경사 하강법은 매 에포크마다 훈련 세트의 샘플 순서를 섞어 가중치 최적값을 계산해야 함.<br>
이는 훈련 세트의 샘플 순서를 섞으면 가중치 최적값의 탐색 과정이 다양해져 가중치 최적값을 제대로 찾을 수 있기 때문.<br>

훈련 세트 자체를 섞는 것 보다 넘파이 배열의 인덱스를 섞은 후 인덱스 순서대로 샘플을 뽑는 것이 효율적이고 빠름.<br>
이를 위해 np.random.permutation() 함수를 사용해서 진행함

[score() 메서드 추가하기]<br>
정확도를 계산해 주는 score() 메서드를 추가하고 predict() 메서드도 조금 수정함.<br>
score() 메서드는 정확도를 직접 계산할 때 사용했던 np.mean() 함수를 사용함

```python
class SingleLayer(SingleLayer):
    def predict(self, x):
        z = [self.forpass(x_i) for x_i in x]   # 정방향 계산
        return np.array(z) >0                  # 스텝 함수 적용
    
    def score(self, x, y):
        return np.mean(self.predict(x) == y)
```

시그모이드 함수의 출력값은 0~1 사이의 확률 값이고 양성 클래스를 판단하는 기준은 0.5 이상임.<br>
여기서 z가 0보다 크면 시그모디으 함수의 출력값은 0.5보다 크고, z가 0보다 작으면 시그모이드 함수의 출력값은 0.5보다 작음.<br>
따라서 predict() 메서드에는 굳이 시그모이드 함수를 사용하지 않아도 됨.<br>
그래서 predict() 메서드에는 로지스틱 함수를 적용하지 않고 z 값의 크기만 비교하여 반환

<h3> 4.2 단일층 신경망 훈련</h3>

1. 단일층 신경망 훈련하고 정확도 출력하기

```python
layer = SingleLayer()
layer.fit(x_train, y_train)
layer.score(x_test, y_test)
```

에포크마다 훈련 세트를 무작위로 섞어 손실 함수의 값을 줄였기 때문에 성능이 좋아짐

<h1>4.7 사이킷런으로 로지스틱 회귀 수행</h1>

<h3>사이킷런으로 경사 하강법 적용하기</h3>

1. 로지스틱 손실 함수 지정하기

```python
from sklearn.linear_model import SGDClassifier
sgd = SGDClassifier(loss='log', max_iter = 100, tol=1e-3, random_state=42)
```

max_iter를 통해 반복 횟수를 100으로 지정.<br>
반복 실행했을 때 결과를 동일하게 재현하기 위해 random_state를 통해 난수 초기값 설정.<br>
반복할 때 마다 로지스틱 손실 함수의 값이 tol에 지정한 값만큼 감소되지 않으면 반복을 중단 설정.<br>

2. 사이킷런으로 훈련하고 평가하기

```python
sgd.fit(x_train, y_train)
sgd.score(x_test, y_test)
```

사이킷 런에는 우리가 여태 구현한 메서드가 이미 준비되어 있음.<br>
fit() 메서드로 훈련하고 score() 메서드로 정확도를 계산함

3. 사이킷런으로 예측하기

```python
sgd.predict(x_test[0:10])
```

마찬가지로 SGDClassifier 클래스에는 예측을 위한 predict() 메서드도 구현되어 있음.<br>
주의할 점은 사이킷런은 데이터로 2차원 배열만 받아들이는 점임.<br>
즉, 샘플 하나를 주입하더라도 2차원 배열로 만들어서 넣어야 함.
