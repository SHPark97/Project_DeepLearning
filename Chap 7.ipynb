```python
# 이전에 필요한 코드들
import numpy as np

class MinibatchNetwork:
    def __init__(self, units=10, batch_size=32, learning_rate=0.1, l1=0, l2=0):
        self.units = units        # 은닉층 뉴런 개수
        self.w1 = None            # 은닉층 가중치
        self.b1 = None            # 은닉층 절편
        self.w2 = None            # 출력층 가중치
        self.b2 = None            # 출력층 절편
        self.a1 = None            # 은닉층의 활성화 출력
        self.losses = []
        self.val_losses = []
        self.lr = learning_rate
        self.l1 = l1
        self.l2 = l2
        self.batch_size = batch_size
        
    def predict(self, x) :
        z = self.forpass(x)
        return z > 0
    
    def update_val_loss(self, x_val, y_val):
        z = self.forpass(x_val)
        a = self.activation(z)
        a = np.clip(a, 1e-10, 1-1e-10)
        val_loss = np.sum(-(y_val*np.log(a) + (1-y_val)*np.log(1-a)))
        self.val_losses.append((val_loss + self.reg_loss()) / len(y_val))
        
    def init_weights(self, n_features):
        np.random.seed(42)
        self.w1 = np.random.normal(0, 1, (n_features, self.units))
        self.b1 = np.zeros(self.units)
        self.w2 = np.random.normal(0, 1, (self.units, 1))
        self.b2 = 0
        
    def forpass(self, x):
        z1 = np.dot(x, self.w1) + self.b1
        self.a1 = self.activation(z1)
        z2 = np.dot(self.a1, self.w2) + self.b2
        return z2
        
    def training(self, x, y, m):
        z = self.forpass(x)
        a =  self.activation(z)
        err = -(y-a)
        
        w1_grad, b1_grad, w2_grad, b2_grad = self.backprop(x, err)
        
        w1_grad += (self.l1 * np.sign(self.w1) + self.l2 * self.w1) / m
        w2_grad += (self.l1 * np.sign(self.w2) + self.l2 * self.w2) / m
        
        self.w1 -= self.lr * w1_grad
        self.b1 -= self.lr * b1_grad
        
        self.w2 -= self.lr * w2_grad
        self.b2 -= self.lr * b2_grad
        return a
    
    def score(self, x, y):
        return np.mean(self.predict(x) == y.reshape(-1, 1))
        
    def reg_loss(self):
        return self.l1 * (np.sum(np.abs(self.w1)) + np.sum(np.abs(self.w2))) + \
                self.l2 / 2 * (np.sum(self.w1**2) + np.sum(self.w2**2))
        
    def fit(self, x, y, epochs=100, x_val=None, y_val=None):
        y_val = y_val.reshape(-1, 1)
        self.init_weights(x.shape[1])
        np.random.seed(42)
        for i in range(epochs):
            loss = 0
            for x_batch, y_batch in self.gen_batch(x, y):
                y_batch = y_batch.reshape(-1, 1)
                m = len(x_batch)
                a = self.training(x_batch, y_batch, m)
                a = np.clip(a, 1e-10, 1-1e-10)
                loss += np.sum(-(y_batch*np.log(a) + (1-y_batch)*np.log(1-a)))
            self.losses.append((loss + self.reg_loss()) / len(x))
            self.update_val_loss(x_val, y_val)
            
    def backprop(self, x, err):
        m = len(x)             # 샘플 개수
        w2_grad = np.dot(self.a1.T, err) / m
        b2_grad = np.sum(err) / m
        # 시그모이드 함수까지 그라디언트 계산
        err_to_hidden = np.dot(err, self.w2.T) * self.a1 * (1 - self.a1)
        # 은닉층의 가중치와 절편에 대한 그라디언트 계산
        w1_grad = np.dot(x.T, err_to_hidden) / m
        b1_grad = np.sum(err_to_hidden, axis=0) / m
        return w1_grad, b1_grad, w2_grad, b2_grad
            
    def gen_batch(self, x, y):
        length = len(x)
        bins = length // self.batch_size
        if length % self.batch_size:
            bins += 1                  # 나누어 떨어지지 않을 때
        indexes = np.random.permutation(np.arange(len(x)))   # 인덱스를 섞음
        x = x[indexes]
        y = y[indexes]
        for i in range(bins):
            start = self.batch_size * i
            end = self.batch_size * (i + 1)
            yield x[start:end], y[start:end]    # batch_size 만큼 슬라이싱하여 반환
```

<h3>7.1.3 크로스 엔트로피 손실 함수를 도입함</h3>

다중 분류의 손실 함수는 로지스틱 손실 함수의 '일반화 버전'인 크로스 엔트로피 손실 함수를 사용함.<br>



크로스 엔트로피 손실 함수의 시그마 기호 위의 값(c)은 전체 클래스 개수를 의미함.<br>
클래스마다 타깃과 출력의 로그값을 곱하여 더한 것임.<br>
분류 문제에서 정답 클래스의 타깃은 1이고 나머지 클래스의 타깃은 0임.<br>
앞의 예제를 적용하면, -1(1xlog0.74 + 0xlog0.18 + 0xlog0.08)이며, <br>
결국 정답 클래스를 제외한 나머지 클래스(y=0)에 대한 손실 항은 모두 소거되어,<br>

이 나오게 됨.<br>

1. 로지스틱 손실 함수와 크로스 엔트로피 손실 함수는 매우 비슷함.<br>

타깃이 양성 클래스(y=1)인 경우와 음성 클래스(y=0)인 경우로 나눠 식을 정리해 봄.<br>
로지스틱 손실 함수는 양성 클래스일 때는 두 번재 항이 소거되고, 음성 클래스일때는 첫 번째 항이 소거됨.<br>
(양성클래스) y = -log a<br>
(음성클래스) y = -log(1-a)<br>

1-a를 음성 클래스의 활성화 출력이라 생각하면 크로스 엔트로피 손실 함수를 정리한 식과 같음을 알 수 있음.<br>
로지스틱 손실 함수는 크로스 엔트로피 손실 함수의 합(시그마) 기호를 빼고 양성 클래스와 음성 클래스에 대해 나누어 놓은 것임.<br>

<h3>7.1.5 다중 분류 신경망 구현</h3>

MinibatchNetwork 클래스를 확장하여 다중 분류를 수행할 수 있는 MultiClassNetwork 클래스를 구현해 봄.<br>

1. 소프트맥스 함수 추가하기<br>

다중 분류에서는 마지막 출력층에 소프트맥스 함수를 사용해야 하므로 은닉층과 출력층에 각기 다른 활성화 함수를 적용.<br>
이를 위해 activation() 메서드의 이름을 sigmoid()로 바꾸고 softmax() 메서드를 추가함.

```python
class MultiClassNetwork(MinibatchNetwork):
    def sigmoid(self, z):
        z = np.clip(z, -100, None)
        a = 1 / (1 + np.exp(-z))
        return a
    
    def softmax(self, z):
        z = np.clip(z, -100, None)
        exp_z = np.exp(z)
        return exp_z / np.sum(exp_z, axis=1).reshape(-1, 1)
```

소프트맥스 함수는 먼저 e^z를 계산해야 함.<br>
np.exp(z)는 z의 각 원소를 e^z로 만들어 줌.<br>



그 후 np.sum() 함수를 사용해 z의 각 행의 합을 계산함.<br>
이때 결과의 형태는 행 벡터(1차원 배열)이므로 형태를 다시 열 벡터(2차원 배열)로 바꿔 나눗셈에 적용할 수 있음.<br>



마지막으로 나눗셈을 적용하면 소프트맥스 함수의 적용이 끝남.



2. 정방향 계산하기<br>

activation() 메서드의 이름을 sigmoid()로 바꿧으니 forward() 메서드에 사용된 activation() 메서드의 이름도 sigmoid()로 바꾸어 줌.

```python
class MultiClassNetwork(MultiClassNetwork):
    def forpass(self, x):
        z1 = np.dot(x, self.w1) + self.b1
        self.a1 = self.sigmoid(z1)
        z2 = np.dot(self.a1, self.w2) + self.b2
        return z2
```

3. 가중치 초기화하기<br>

이진 분류에서는 출력층의 뉴런이 1개이므로 가중치 w2의 크기는 (은닉층의 뉴런 개수, 1) 로 지정함.<br>
다중 분류에서는 출력층의 뉴런이 2개 이상이므로 가중치 w2의 크기는 (은닉층의 뉴런 개수, 클래스 개수) 로 지정함.<br>
b2의 크기는 클래스 개수에 따라 지정함.<br>
w2는 random.normal() 함수를 이용하여 배열의 각 원소의 값을 정규 분포를 따르는 무작위 수로 초기화함.<br>
b2는 모두 0으로 초기화 함.

```python
class MultiClassNetwork(MultiClassNetwork):
    def init_weights(self, n_features, n_classes):
        np.random.seed(42)
        self.w1 = np.random.normal(0, 1, (n_features, self.units))
        self.b1 = np.zeros(self.units)
        self.w2 = np.random.normal(0, 1, (self.units, n_classes))
        self.b2 = np.zeros(n_classes)
```

4. fit 메서드 수정하기<br>

가중치를 초기화하는 init_weights() 메서드를 호출할 때 클래스의 개수를 매개변수의 값으로 넘겨줌.<br>
다중 분류 문제에서 y, y_val은 2차원 행렬이므로 열 벡터로 변환하던 코드를 지움.<br>
추가로 편의상 훈련의 진행 상황을 살펴볼 수 있도록 print() 함수를 사용하여 에포크마다 '.'이 출력되도록 함.

```python
class MultiClassNetwork(MultiClassNetwork):
    def fit(self, x, y, epochs=100, x_val=None, y_val=None):
        np.random.seed(42)
        self.init_weights(x.shape[1], y.shape[1])  # 은닉층과 출력층의 가중치 초기화
        for i in range(epochs):
            loss = 0
            print('.', end='')
            for x_batch, y_batch in self.gen_batch(x, y):
                a = self.training(x_batch, y_batch)
                a = np.clip(a, 1e-10, 1-1e-10)
                loss += np.sum(-y_batch*np.log(a))
            self.losses.append((loss + self.reg_loss()) / len(x))
            self.update_val_loss(x_val, y_val)
```

5. training() 메서드 수정하기 <br>

training() 메서드에서 사용하는 출력층의 호라성화 함수를 activation() 메서드에서 softmax() 메서드로 변경

```python
class MultiClassNetwork(MultiClassNetwork):
    def training(self, x, y):
        m = len(x)
        z = self.forpass(x)
        a =  self.softmax(z)
        err = -(y-a)
        
        w1_grad, b1_grad, w2_grad, b2_grad = self.backprop(x, err)
        
        w1_grad += (self.l1 * np.sign(self.w1) + self.l2 * self.w1) / m
        w2_grad += (self.l1 * np.sign(self.w2) + self.l2 * self.w2) / m
        
        self.w1 -= self.lr * w1_grad
        self.b1 -= self.lr * b1_grad
        
        self.w2 -= self.lr * w2_grad
        self.b2 -= self.lr * b2_grad
        return a
```

6. predict() 메서드 수정하기<br>

predict() 메서드에서는 정방향 계산에서 얻은 출력 중 가장 큰 값의 인덱스를 구함.<br>
이 값이 예측 클래스가 됨.<br>
이진 분류에서와 마찬가지로 클래스를 예측할 때는 활성화 함수, 즉 소프트맥스 함수를 거칠 필요가 없음.<br>
출력층에서 계산된 선형 계산만으로 클래스를 예측할 수 있음.

```python
class MultiClassNetwork(MultiClassNetwork):
    def predict(self, x):
        z = self.forpass(x)
        return np.argmax(z, axis=1)
```

7. score() 메서드 수정하기<br>

score() 메서드에서는 predict() 메서드의 결과와 타깃 y의 클래스를 비교함.<br>
이를 위해 배열 y의 행을 따라 가장 큰 값의 인덱스를 구해 사용함.

```python
class MultiClassNetwork(MultiClassNetwork):
    def score(self, x, y):
        return np.mean(self.predict(x) == np.argmax(y, axis=1))
```

8. 검증 손실 계산하기<br>

update_val_loss() 메서드에서 사용하는 활성화 함수를 softmax()로 바꿈.<br>
또 로지스틱 손실 계산을 크로스 엔트로피 손실 계산으로 바꿈.

```python
class MultiClassNetwork(MultiClassNetwork):
    def update_val_loss(self, x_val, y_val):
        z = self.forpass(x_val)
        a = self.softmax(z)
        a = np.clip(a, 1e-10, 1-1e-10)
        val_loss = np.sum(-y_val*np.log(a))
        self.val_losses.append((val_loss + self.reg_loss()) / len(y_val))
```

위 그림을 보며 원-핫 인코딩을 설명함.<br>
예를 들어 타깃값 6은 인덱스 6에 해당하는 원소만 1이고 나머지 원소는 0이 됨.<br>

<h3>7.2.1 케라스에 대해 알아봄</h3>

케라스는 딥러닝 패키지를 편리하게 사용하기 위해 만들어진 래퍼 패키지임.<br>
아래는 간단한 신경망을 케라스로 구현한 것임.<br>

```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense

# 신경망 모델을 만듦
model = tf.keras.models.Sequential()
# 완전 연결층을 추가함
model.add(tf.keras.layers.Dense(1))
# 옵티마이저와 손실 함수 지정
model.compile(optimizer='sgd', loss='mse')
# 훈련 데이터를 사용하여 에포크 횟수만큼 훈련함
model.fit(x_train, y_train, epochs=10)
```

<li>케라스를 사용하면 인공신경망의 층을 직관적으로 설계할 수 있음</li>



예를 들어 위와 같은 다중 분류 신경망을 만든다고 생각해보자.<br>
케라스는 인공신경망 모델을 만들기 위한 Sequential 클래스와 완전 연결층을 만들기 위한 Dense 클래스를 제공함.<br>

Sequential 클래스는 이름의 의미 그대로 '순차적으로 층을 쌓은 신경망 모델'이고,<br>
Dense 클래스는 모델에 포함된 완전 연결층임.<br>
즉, 위의 인공신경망은 은닉층과 출력층을 Dense 클래스의 객체로 구성하고 각각의 객체를 Sequential 클래스 객체에 추가해 완성됨.<br>
이는 정말 층을 쌓아 신경망을 만든 것처럼 보이므로 '직관적으로 인공신경망의 층을 설계한다' 고 말함.<br>

우리가 설계한 MultiClassNetwork 클래스의 경우 신경망 모델과 층이 명확하게 구분되지 않음.<br>
하지만 케라스는 모델과 층이 명확하게 구분됨.<br>
심지어 층 하나를 다른 층으로 교체하기도 쉬움.<br>
이 두 클래스를 이용해 다중 분류 신경망을 만들어 봄.

4. 모델 훈련하기<br>

훈련하는 동안 검증 세트에 대한 손실과 정확도를 계산할 수 있음.<br>
이를 위해 validation_data 매개변수에 검증 세트를 튜플로 전달함.<br>
fit() 메서드는 훈련 세트와 검증 세트에서 측정한 값들을 History 클래스 객체에 담아 반환함.<br>
반환값을 history 변수에 저장할 것임.

```python
history = model.fit(x_train, y_train_encoded, epochs=40, validation_data=(x_val, y_val_encoded))
```

5. 손실과 정확도 그래프 그리기<br>

history 객체의 history 딕셔너리에는 여러 측정 지표가 들어 있음.<br>
history 딕셔너리의 키를 출력해 보면 어떤 측정 지표들이 들어 있는지 알 수 있음.<br>

```python
print(history.history.keys())
```

훈련 세트와 검증 세트에서의 손실을 나타내는 loss, val_loss가 있고, <br>
훈련 세트와 검증 세트의 정확도를 의미하는 accuracy와 val_accuracy가 있음.<br>
다음은 훈련 세트와 검증 세트의 손실 그래프와 정확도를 나타낸 그래프임.

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train_loss', 'val_loss'])
plt.show()

plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train_accuracy', 'val_accuracy'])
plt.show()
```

출력된 그래프를 보면 손실 그래프는 일정한 수준으로 감소하는 추세를 보이고 정확도 그래프도 점진적으로 증가하고 있음.<br>
